```python
import pandas as pd
from pathlib import Path
```

```python
jan_data_2023 = Path('../citibike_files/JC-202301-citibike-tripdata.csv')
feb_data_2023 = Path('../citibike_files/JC-202302-citibike-tripdata.csv')
mar_data_2023 = Path('../citibike_files/JC-202303-citibike-tripdata.csv')
apr_data_2023 = Path('../citibike_files/JC-202304-citibike-tripdata.csv')
may_data_2023 = Path('../citibike_files/JC-202305-citibike-tripdata.csv')
jun_data_2023 = Path('../citibike_files/JC-202306-citibike-tripdata.csv')
jul_data_2023 = Path('../citibike_files/JC-202307-citibike-tripdata.csv')
aug_data_2023 = Path('../citibike_files/JC-202308-citibike-tripdata.csv')
sep_data_2023 = Path('../citibike_files/JC-202309-citibike-tripdata.csv')
oct_data_2023 = Path('../citibike_files/JC-202310-citibike-tripdata.csv')
nov_data_2023 = Path('../citibike_files/JC-202311-citibike-tripdata.csv')
dec_data_2023 = Path('../citibike_files/JC-202312-citibike-tripdata.csv')
```

```python
jan_2023 = pd.read_csv(jan_data_2023)
feb_2023 = pd.read_csv(feb_data_2023)
mar_2023 = pd.read_csv(mar_data_2023)
apr_2023 = pd.read_csv(apr_data_2023)
may_2023 = pd.read_csv(may_data_2023)
jun_2023 = pd.read_csv(jun_data_2023)
jul_2023 = pd.read_csv(jul_data_2023)
aug_2023 = pd.read_csv(aug_data_2023)
sep_2023 = pd.read_csv(sep_data_2023)
oct_2023 = pd.read_csv(oct_data_2023)
nov_2023 = pd.read_csv(nov_data_2023)
dec_2023 = pd.read_csv(dec_data_2023)
```

```python
df_2023 = pd.concat([jan_2023, feb_2023, mar_2023, apr_2023, may_2023, jun_2023, jul_2023, aug_2023, sep_2023, oct_2023, nov_2023, dec_2023])
df_2023.reset_index(drop=True, inplace=True)
df_2023.set_index('ride_id', inplace=True)
df_2023
```

```python
df_2023.to_csv('../citibike_files/citibike_2023.csv')
```
